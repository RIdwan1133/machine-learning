```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('kalori_ganda.csv')
```

```python
#1. scatter plot untuk melihat hubungan antara umur dan kalori
plt.figure(figsize=(7,5))
plt.scatter(x='Umur', y='Kalori', data=df, color='blue')
plt.title('Age vs Calorie Intake')
plt.xlabel('Age (years)')
plt.ylabel('Calorie Intake (kcal)')
plt.grid(True)
plt.show()
```

```python
#2 Berat Badan vs Konsumsi Kalori
plt.figure(figsize=(7,5))
plt.scatter(x='BB', y='Kalori', data=df, color='green')
plt.title('Weight vs Calorie Intake')
plt.xlabel('Weight (kg)')
plt.ylabel('Calorie Intake(kcall)')
plt.grid(True)
plt.show()
```

```python
# Histogram untuk melihat distribusi BB
plt.figure(figsize=(8,6))
sns.histplot(x='BB', data=df)
plt.title('Distribusi BB')
plt.show()
```

```python
fig = px.scatter_3d(df, x='Umur', y='BB', z='TB',
                    color='Olahraga', size='Kalori',
                    hover_name='Umur')
fig.show()
```

```python
# Memisahkan fitur (X) dan target (y)
x = df[['Umur', 'BB', 'TB','Olahraga']]
y = df['Kalori']
```

```python
# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

```python
#membuat model regresi linier
model = LinearRegression()
model.fit(x, y)
```

```python
 # Menampilkan intercept regresi dari model
intercept = model.intercept_
intercept
```

```python
#Menampilkan koefisien regresi dari model
coefficients = model.coef_
coefficients
```

```python
# Data input untuk prediksi
input_data =np.array([[25, 75, 170,60]])
```

```python
# Melakukan prediksi 
predicted_calories =model.predict(input_data)
predicted_calories
```

```python
# Memprediksi nilai BB dengan data uji
y_pred = model.predict(X_test)
```

```python
#Menghitung MSE dan R2 score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```
